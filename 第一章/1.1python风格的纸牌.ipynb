getitem&len方法

**魔术方法magic和双下方法dunder**(代表有四个下划线）

```python
import collections
# namedtuple 构建只有少量属性没有方法的对象！！
Card = collections.namedtuple('Card',['rank','suit'])

class FrenchDeck:
    # 类内可以定义局部变量 都默认为self的
    ranks = [str(n) for n in range(2,11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()
    def __init__(self):
        #tql 直接生成二元组 13*4
        self.cards = [Card(rank,suit) for suit in self.suits 
                                      for rank in self.ranks]
        
    def __len__(self):
        return len(self.cards)
    
    def __getitem__(self,position):
        return self.cards[position]
```

```python
deck = FrenchDeck()
len(deck)
```

```python
deck[-1]
```

```python
#随机选取元素 只要内置了len&getitem就能使用
from random import choice
choice(deck)
```

```python
#支持自动切片--position传递给cards了
deck[:3]
```

```python
#可迭代了！ reversed是反向！
#也可以进行 in 操作 尽管没有实现__contains__方法 会全部迭代遍历一遍
for card in reversed(deck): # doctest: +ELLIPSIS
    print(card)
```

```python
# 排序 要定义排序函数

# 字典初始化可以字符串可以直接写字母
suit_values = dict(spades=3,hearts=2,diamonds=1,clubs=0)

#print(suit_values['spades'])

def spades_high(card):
    #.index 返回索引位置
    rank_value = FrenchDeck.ranks.index(card.rank)
    
    #让数字占排序主导地位
    return rank_value*10 + suit_values[card.suit]
```

```python
for card in sorted(deck,key=spades_high):
    print(card)
```

1.小明设计了一种手机产品，有手机大小，手机重量，手机价格三个属性。每个属性都分为1~10等级。请将所有可能的手机类型作为一个类成员记录在手机类中，并构造一些函数使其可以切片，遍历，并实现按照大小重量价格依次降序来排序。

```python
import collections
Phone = collections.namedtuple('Phone',['size','weight','price'])

class Phones:
    sizes=[str(i) for i in range(1,11)]
    weights=[str(i) for i in range(1,11)]
    prices=[str(i) for i in range(1,11)]
    def __init__(self):
        self.phones = [Phone(size,weight,price) for size in self.sizes
                                                for weight in self.weights
                                                for price in self.prices]
    def __len__(self):
        return len(self.phones)
    
    def __getitem__(self,p):
        return self.phones[p]
```

```python
phone1=Phones()
```

```python
len(phone1)
```

```python
for phone in reversed(phone1[:5]):
    print(phone)
```

```python
from random import choice 
choice(phone1)
```

```python
def sort_value(phone):
    return Phones.prices.index(phone.price)*100+Phones.weights.index(phone.weight)*10+Phones.sizes.index(phone.size)

for i in sorted(phone1,key=sort_value,reverse=True):
    print(i)
```
